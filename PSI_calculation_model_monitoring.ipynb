```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#from psi import calculate_psi
```

```python
sns.set_style("white")
rs = np.random.RandomState(5)
```

```python
initial = rs.normal(size = 100)
initial
```

```python
new = rs.normal(loc = 0.2, size = 120)
```

```python
new
```

```python
def scale_range (input, min, max):
    input += -(np.min(input))
    input /= np.max(input) / (max - min)
    input += min
    return input
buckets = 10
raw_breakpoints = np.arange(0, buckets + 1) / (buckets) * 100
breakpoints = scale_range(raw_breakpoints, np.min(initial), np.max(initial))
```

```python
breakpoints
```

```python
initial_counts = np.histogram(initial, breakpoints)[0]
new_counts = np.histogram(new, breakpoints)[0]
print(initial_counts)
print(new_counts)
```

```python
df = pd.DataFrame({'Bucket': np.arange(1, 11), 'Breakpoint Value':breakpoints[1:], 'Initial Count':initial_counts, 'New Count':new_counts})
```

```python
df
```

```python
df['Initial Percent'] = df['Initial Count'] / len(initial)
df['New Percent'] = df['New Count'] / len(new)
```

```python
df['New Percent'][df['New Percent'] == 0] = 0.001
```

```python
df
```

```python

df['PSI'] = (df['New Percent'] - df['Initial Percent']) * np.log(df['New Percent'] / df['Initial Percent'])
```

```python
df
```

```python
np.sum(df['PSI'])
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/deepanshu88/data/master/data.csv")
```

```python
df['y0']=1-df['y']
df['bucket']=pd.qcut(df['p'],10)
grouped=df.groupby('bucket',as_index=False)
kstable=pd.DataFrame()
```

```python
df.groupby('bucket',as_index=False).head()
```

```python
type(grouped)
```

```python
kstable['min_prob'] = grouped.min()['p']
kstable['max_prob'] = grouped.max()['p']
kstable['events']   = grouped.sum()['y']
kstable['nonevents'] = grouped.sum()['y0']
kstable = kstable.sort_values(by="min_prob", ascending=False).reset_index(drop = True)
kstable['event_rate'] = (kstable.events / df['y'].sum()).apply('{0:.2%}'.format)
kstable['nonevent_rate'] = (kstable.nonevents / df['y0'].sum()).apply('{0:.2%}'.format)
kstable['cum_eventrate']=(kstable.events / df['y'].sum()).cumsum()
kstable['cum_noneventrate']=(kstable.nonevents / df['y0'].sum()).cumsum()
kstable['KS'] = np.round(kstable['cum_eventrate']-kstable['cum_noneventrate'], 3) * 100
```

```python
# Cramer's V calculation.

from itertools import product
import numpy as np
import pandas as pd
import scipy.stats as stats

np.random.seed(42)

all_combs = product(
    ['Spring', 'Summer', 'Fall', 'Winter'],
    ['New', 'Old'],
    ['Cold', 'Warm', 'Hot']
)

df = pd.DataFrame(all_combs, columns=['Season', 'Age', 'Weather'])
df['Sales'] = np.random.randint(25, 200, len(df))
df.head()
```

```python
cont = df.pivot_table('Sales', 'Season', 'Weather', 'sum')
cont
```

```python
cont_arr = cont.values
```

```python
cont_arr
```

```python
chi2 = stats.chi2_contingency(cont_arr, correction=False)[0]
sample_size = np.sum(cont_arr)
min_dim = min(cont_arr.shape) - 1

cramer_v = np.sqrt((chi2 / sample_size) / min_dim)

cramer_v
```
